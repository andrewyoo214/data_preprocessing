### Using seaborn library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#tips라는 데이터를 불러와서 앞으로 써보자
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Drawing historgram with seaborn library
a = plt.subplots() #set the figure
a = sns.distplot(tips['total_bill'])
a.set_title('Total Bill Histogram with Density Plot')
```

```python
# what if we do not want the density distribution plot?
a = plt.subplots() #set the figure
a = sns.distplot(tips['total_bill'], kde=False)
a.set_title('Total Bill Histogram')
a.set_xlabel('Total Bill')
a.set_ylabel('Frequency')
```

```python
# what if we only want to see the density distribution plot?
a = plt.subplots() #set the figure
a = sns.distplot(tips['total_bill'], hist=False)
a.set_title('Total Bill Histogram')
a.set_xlabel('Total Bill')
a.set_ylabel('Unit Probability')
```

```python
# Normally we put them altogether with rug plot to efficiently show the data
a = plt.subplots()
a = sns.distplot(tips['total_bill'], rug=True)
a.set_title('Total Bill Histogram')
a.set_xlabel('Total Bill')
```

```python
# Drawing count graph
a = plt.subplots() #set the figure
a = sns.countplot('day', data=tips)
a.set_title('Count of Days')
a.set_xlabel('Day of the week')
a.set_ylabel('Frequency')
```

##### Bivariate Graphs

```python
# Drawing Scatterplot with seaborn
a = plt.subplots() #set the figure
a = sns.regplot(x = 'total_bill', y='tip', data = tips)
a.set_title('Scatterplot of Total Bill and Tip')
a.set_xlabel('Total Bill')
a.set_ylabel('Tip')
```

```python
# Scatter plot without fitted regression line
a = plt.subplots() #set the figure
a = sns.regplot(x = 'total_bill', y='tip', data = tips, fit_reg=False)
a.set_title('Scatterplot of Total Bill and Tip')
a.set_xlabel('Total Bill')
a.set_ylabel('Tip')
```

```python
# What if we want to see both scatter plot and histogram of our data?
b = sns.jointplot(x = 'total_bill', y='tip', data = tips)
b.set_axis_labels(xlabel = 'Total Bill', ylabel='Tip')
b.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize = 10, y = 1.03)
```

```python
# When we have huge amount of data, scatter plot has trouble with showing overlapped dataset
# we can use hexagon plot to figure out the problem
h = sns.jointplot(x = 'total_bill', y = 'tip', data=tips, kind = "hex")
h.set_axis_labels(xlabel = 'Total Bill', ylabel='Tip')
h.fig.suptitle('Hexagon Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
# Drawing 2D density plot
a = plt.subplots()
a = sns.kdeplot(data=tips['total_bill'],
               data2 = tips['tip'],
               shade=True)
a.set_title('Kernel Density Plot of Total Bill and Tips')
a.set_xlabel('Total Bill')
a.set_ylabel('Tip')
```

```python
# Drawing Bar Plot
a = plt.subplots()
a = sns.barplot(x='time', y='total_bill',data=tips)
a.set_title('Bar plot of average total bill for time of day')
a.set_xlabel('Time of the day')
a.set_ylabel('Average total bill')
```

```python
# Drawing a Box Plot
# Boxplot is useful to check the statistical summaries
a = plt.subplots()
a = sns.boxplot(x='time', y='total_bill',data=tips)
a.set_title('Box plot of Total Bill by time of day')
a.set_xlabel('Time of the day')
a.set_ylabel('Total bill')
```

```python
# Better way to check boxplot?
# Box plot + Kernel density plot
# violinplot, because it seems like a violin
# It shows the distribution of the data, like population distribution data
a = plt.subplots()
a = sns.violinplot(x='time', y='total_bill',data=tips)
a.set_title('Violin plot of Total Bill by time of day')
a.set_xlabel('Time of the day')
a.set_ylabel('Total bill')
```

```python
# Drawing a pair Relationship Graph
fig = sns.pairplot(tips)
```

```python
pair = sns.PairGrid(tips)
pair = pair.map_upper(sns.regplot)
pair = pair.map_lower(sns.kdeplot)
pair = pair.map_diag(sns.distplot, rug = True)
plt.show()
```

##### Drawing a Multivariate Graph

```python
# Violin graph with color seperation
a = plt.subplots()
a = sns.violinplot(x='time', y='total_bill', hue = 'sex', data = tips, split=True)
# hue='sex' option to seperate the colors between male and female
```

```python
# Scatterplot and pairplot with color seperation
a = sns.lmplot(x='total_bill', y = 'tip', data=tips, hue='sex', fit_reg = False)
fig = sns.pairplot(tips, hue='sex')
```

```python
# Applying different elements on scatterplot
a = sns.lmplot(x='total_bill', y = 'tip', data=tips, hue='sex', fit_reg = False, markers=['o','x'])
```

```python
anscombe = sns.load_dataset("anscombe")
a = sns.lmplot(x='x', y='y', data=anscombe, fit_reg=False)
```

```python
# Drawing a graph for each group
a_plot = sns.lmplot(x='x', y='y', data=anscombe, fit_reg = False, col='dataset', col_wrap=2)
```

##### Applying FacetGrid class to draw grouped graphs

```python
c = sns.FacetGrid(tips, col='time')
c.map(sns.distplot, 'total_bill', rug=True)
```

```python
c = sns.FacetGrid(tips, col='day', hue='sex')
c = c.map(plt.scatter, 'total_bill', 'tip')
c = c.add_legend()
```

```python
c = sns.FacetGrid(tips, col='time',row='smoker',hue='sex')
c.map(plt.scatter, 'total_bill', 'tip')
```

##### Drawing a graph with Dataframe and Pandas Series

```python
# histogram with hist method in plot lib.
a = plt.subplots()
a = tips['total_bill'].plot.hist()
```

```python
# Adjusting transparency on the histogram. and the x axis length
fig, a = plt.subplots()
a = tips[['total_bill', 'tip']].plot.hist(alpha=0.5, bins=20, ax=a)
#alpha changes transparency, bins changes length between variables
```

```python
# kernel density plot, scatterplot, hexagon plot
# kde, scatter, hexbin
a = plt.subplots()
a = tips['tip'].plot.kde()
fig, a = plt.subplots()
a = tips.plot.scatter(x = 'total_bill', y = 'tip', ax=a)
fig, a = plt.subplots()
a = tips.plot.hexbin(x='total_bill', y= 'tip', ax=a)
```

```python
# we can change grid size of the hexagon plot with gridsize option
fig, a = plt.subplots()
a = tips.plot.hexbin(x='total_bill', y='tip', gridsize=10, ax=a)
```

```python
# lets draw box plot
fig,a = plt.subplots()
a = tips.plot.box(ax=a)
```

##### Setting a graph style with seaborn library

```python
# Drawing a violin plot
sns.set_style('whitegrid') #put the whitegrid on the background
fig, a = plt.subplots()
a = sns.violinplot(x='time',y='total_bill', hue= 'sex', data=tips, split=True)
```

```python
# using 'for' to applying every styles that we can apply
fig = plt.figure()

sns_styles=['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

for idx, style in enumerate(sns_styles):
    plot_position = idx + 1
    with sns.axes_style(style):
        a = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=a)
        violin.set_title(style)
fig.tight_layout()
```
