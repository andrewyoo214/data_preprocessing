### > Data Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Anscombe = graph to show that the error of data checking withouth visualization methods
anscombe = sns.load_dataset("anscombe")
anscombe.head()
```

```python
dataset1 = anscombe[anscombe['dataset']== 'I'] #data 값이 I 에 해당하는 것만 추출
plt.plot(dataset1['x'], dataset1['y']) # 선마다 이어진 상태로 그려짐
```

```python
# 점으로 scatter plot 형태로 나타내려면 'o' method를 인자로 전달하자
plt.plot(dataset1['x'], dataset1['y'], 'o')
```

```python
dataset2 = anscombe[anscombe['dataset']== 'II']
dataset3 = anscombe[anscombe['dataset']== 'III']
dataset4 = anscombe[anscombe['dataset']== 'IV']
```

```python
# 그림을 그릴 배경을 만들어주는 것
fig = plt.figure(figsize=(10,10))
# 정의한 figure에 대해서 위치를 잡아준다
a1 = fig.add_subplot(2,2,1)
a2 = fig.add_subplot(2,2,2)
a3 = fig.add_subplot(2,2,3)
a4 = fig.add_subplot(2,2,4)

# 위에서 정의한 데이터셋을 이용하여 plot을 그려보자
a1.plot(dataset1['x'], dataset1['y'], 'o')
a2.plot(dataset2['x'], dataset2['y'], 'o')
a3.plot(dataset3['x'], dataset3['y'], 'o')
a4.plot(dataset4['x'], dataset4['y'], 'o')

# 이름을 넣어 좀 더 잘 보이게 한다
a1.set_title("dataset #1")
a2.set_title("dataset #2")
a3.set_title("dataset #3")
a4.set_title("dataset #4")

# 전체 figure에 대한 제목도 달아준다
fig.suptitle("Anscombe Data")
```

#### matplotlib 을 활용해서 visualization위한 graph 만들기

```python
#tips라는 데이터를 불러와서 앞으로 써보자
tips = sns.load_dataset("tips")
tips.head()
```

```python
# 기본적인 figure틀을 만들고 틀 짜는건 동일하게 가면 됨.
fig = plt.figure()
a1 = fig.add_subplot(1, 1, 1)

# tips 데이터를 활용해서 histogram을 그려보자
a1.hist(tips['total_bill'], bins=10)
a1.set_title('Histogram of Total Bill')
a1.set_xlabel('Frequency')
a1.set_ylabel('Total Bill')
```

```python
# 기본적인 figure틀을 만들고 틀 짜는건 동일하게 가면 됨.
fig = plt.figure()
a1 = fig.add_subplot(1, 1, 1)

# tips 데이터를 활용해서 scatter plot을 그려보자
a1.scatter(tips['total_bill'], tips['tip'])
a1.set_title('Scatter plot of Total Bill vs Tip')
a1.set_xlabel('Total Bill')
a1.set_ylabel('Tip')
```

```python
# 기본적인 figure틀을 만들고 틀 짜는건 동일하게 가면 됨.
fig = plt.figure()
a1 = fig.add_subplot(1, 1, 1)

a1.boxplot([tips[tips['sex'] == 'Female']['tip'],
           tips[tips['sex'] == 'Male']['tip']],
          labels = ['Female', 'Male'])

a1.set_xlabel('Sex')
a1.set_ylabel('Tip')
a1.set_title('Boxplot of Tips by Sex')
```

```python
# 다변량 데이터를 통한 다변량 그래프

# 여성과 남성에 대해 정수로 반환해줄 수 있는 함수
def encode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1

#위에서 encode하는 함수를 만들어놓았으니 apply 해보자
tips['sex_color'] = tips['sex'].apply(encode_sex)
```

```python
# 성별별로 산점도를 그려보자
scatterplot = plt.figure()
a1 = scatterplot.add_subplot(1, 1, 1)
# 산점도에 그려줄 항목들을 표기함
a1.scatter(x = tips['total_bill'],
          y = tips['tip'],
          s = tips['size'] * 10,
          c = tips['sex_color'],
          alpha = 0.5)

# 타이틀을 달아주자
a1.set_title('Total Bill vs Tip colored by sex and sized')
a1.set_xlabel('Total Bill')
a1.set_ylabel('Tip')
```
